```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading the audio file
audio_file = "/home/redeus/Documents/INFO L3/S2/INF 372/EC1/Datasets/audio/chiffre_01.wav"
```

```python
ipd.Audio(audio_file)
```

```python
signal, sr = librosa.load(audio_file)
```

```python
signal.shape
```

## Extractinf MFCC

```python
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=13, sr=sr)
```

```python
mfccs.shape
```

## Visualise the MFCCS

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, x_axis="time", sr = sr)
plt.colorbar(format="%+2f")
plt.show()
```

## Calculating the the derivatives of the MFCC

```python
delta_mfcc = librosa.feature.delta(mfccs)
delta2_mfcc = librosa.feature.delta(mfccs, order = 2)
```

```python
delta_mfcc.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfcc, x_axis="time", sr = sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfcc, x_axis="time", sr = sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
comprehensive_mfccs = np.concatenate((mfccs, delta_mfcc, delta2_mfcc))
```
